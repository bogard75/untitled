```python
from selenium import webdriver

# Geckodriver 실행본 파일 locate
browser = webdriver.Firefox(executable_path='/home/bogard75/다운로드/geckodriver')
# Selenium Test : manaa.space
browser.get("http://manaa.space/comics/?mode=1&order=3&status=2")

html = browser.page_source
soup = BeautifulSoup(html, 'html.parser')
k1 = soup.findAll('div', attrs={"class": "col-lg-4 col-md-4 col-md-6 col-xs-12"})
comics = soup.findAll('a', attrs={"class":"nothumb-comics-item"})
```

```python
# 제목 리스트 (제목, 링크)
comics = [(c.text, c['href']) for c in comics]
```

```python
# 만화책 리스트 (링크, 권, 일자)
def books(comic) :
    subj = comic[0]
    addr = 'http://manaa.space%s' % comic[1]
    
    browser.get(addr)
    html = browser.page_source
    soup = BeautifulSoup(html, 'html.parser')
    link = soup.findAll('a', attrs={"class": "style-scope comics-post-list"})
    num = soup.findAll('h4', attrs={"class": "classification style-scope comics-post-list"})
    date = soup.findAll('span', attrs={"class": "date style-scope comics-post-list"})
    books = [(subj, l['href'], n.text, d.text) for l, n, d in zip(link, num, date)]
    return books

# 전체 제목의 만화책 리스트 담기
book = list(map(books, comics))
```
